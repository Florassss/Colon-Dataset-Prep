```python
def reinhard_img(directory, des_path):
    img = cv2.imread('dataset/new_crypts/train/Images/1005474.svs (1, 8456, 15758, 1299, 823).png')
    mean_ref, std_ref = htk.preprocessing.color_conversion.lab_mean_std(img)
    
    for filename in os.listdir(directory):
        print(filename)
        if filename.endswith(".png"): 
            img_input = cv2.imread(directory+filename)
            
            im_nmzd = htk.preprocessing.color_normalization.reinhard(img_input, mean_ref, std_ref)
            
            status = cv2.imwrite(os.path.join(des_path)+filename, im_nmzd)
            print("Image written to file-system : ",status)
            continue
        else:
            continue
```

```python
reinhard_img('dataset/new_crypts/valid/Images/', 'dataset/new_crypts/valid/normalized_imgs')
```
